## Pengenalan pyplot
* pyploy merupakan koneksi atau kummpulan fungsi yang menjadikan Matplotlib dapat bekerjan seperti MATLAB.


* Pyplot Api secara umum less-flexsible bila dibandingkan dengan OO API.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print (matplotlib.__version__)
print (np.__version__)
```

## membuat ploting sederhana

```python
plt.plot([2,5,7,11])
plt.ylabel('sumbu y')
plt.show()
```

```python
plt.plot([1,2,3,4,5])
plt.show()
```

```python
plt.plot([1,2,3,4,5],[5,4,9,2,1])
plt.show()
```

## pengaturan formating pada plot

```python
# karakter utama untuk warna "g" dan karakter ke dua untuk simbil "o"
plt.plot([1,2,3,4],[1,4,9,16], 'go')

# [xmin, xmax, ymin, ymax]
plt.axis([0,8,0,20])

plt.show()
```

## multi plot dalam satu pemanggilan fungsi plot()


```python
# parameter ketiga untuk mengatur interval pada deret bilangan.
t = np.arange(0.,5.,0.2)
t
```

```python
# garis putus - putus berwarna biru
plt.plot(t, t, 'r--')
# data kuadrat warna biru "b" dan kotak "s"
plt.plot(t,t**2, 'bs')
# dapa kubik wana green "g" dan caping "^"
plt.plot(t,t**3, 'g^')

plt.show()
```

## meringkas plot

```python
plt.plot(
    t,t,'r--',
    t,t**2, 'bs',
    t,t**3, 'g^'
)
plt.show()
```

## ploting dengan keyword

```python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0,50,50),
    'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d'])*100

data
```

```python
plt.scatter('a','b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# "c" untuk warna dan "s" untuk ukuran
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## ploting untuk tipe data categorical

```python
name = ['group_a', 'group_b', 'group_c']
value = [1,10,100]
```

```python
plt.plot(name, value)
plt.show()
```

```python
plt.scatter(name, value)
plt.show()
```
