## mengenal Objek oriented style untuk melakukan ploting

secara eksplisit membuat figure beserta exes, dan memanggil method dari keduanya

```python
x = np.linspace(0,2,100)
x
```

```python
fig, ax = plt.subplots() # membuah figure

# ploting tiga variable data pada axes
ax.plot(x,x, label='linear')
ax.plot(x,x**2, label='quadrat')
ax.plot(x,x**3, label='cubic')

# Mengatur layout figure
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('simple plot')
ax.legend()



```
